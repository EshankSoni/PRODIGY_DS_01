```python
#importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('population data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# dropping unnamed column
df= df.drop("Unnamed: 67", axis = 1)
df.columns
```

```python
df.info()
```

```python
# checking total null values
df.isnull().sum().sum()
```

```python
# dropping null values
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
# chosing the country to visualize the pupulation over the years
Chosen_country = df[df['Country Name'] == 'India']
Chosen_country
```

```python
# bar plot
plt.figure(figsize = (15,8))
sns.barplot(Chosen_country, palette= 'magma_r')
plt.xticks(rotation=90)
plt.title('Population of India over the Years')
plt.xlabel('Years')
plt.ylabel('pupulation total (in Billions)')
plt.show()
```

```python
# choosing one other country to compare its population growth with India
other_country= df[df['Country Name'] == 'China']
other_country
```

```python
plt.figure(figsize = (15,8))
sns.barplot(other_country, palette= "ch:s=.25,rot=-.25")
plt.xticks(rotation=90)
plt.title('Population of China over the Years')
plt.xlabel('Years')
plt.ylabel('pupulation total (in Billions)')
plt.show()
```

```python
# Comparing the population growth of both country
plt.figure(figsize = (15,8))
plt.subplot(211)
sns.barplot(Chosen_country, palette= 'magma_r')
plt.xticks(rotation=90)
plt.title('Population of India over the Years')
plt.xlabel('Years')
plt.ylabel('pupulation total (in Billions)')
plt.show()

plt.figure(figsize = (15,8))
plt.subplot(212)
sns.barplot(other_country, palette= "ch:s=.25,rot=-.25")
plt.xticks(rotation=90)
plt.title('Population of China over the Years')
plt.xlabel('Years')
plt.ylabel('pupulation total (in Billions)')
plt.show()
```
